 Q. 10) What is the maximum duration of a Movie/Show on Netflix ?

```python
df['Duration'].unique
```

```python
# Since Duration consists of min and seasons values, we need to seperate the column using split fun-
df[['Dur_num','Dur_unit']] = df['Duration'].str.split(' ',expand=True)
df.head()
```

```python
def new_dur(X,Y):
    Dur_new=[]
    for (x,y) in zip(X,Y):
        if y =='Seasons':
          Dur_new.append(int(x)*20)
        else:
            Dur_new.append(x)  
    return Dur_new   



df['dur_new']= new_dur(df['Dur_num'],df['Dur_unit'])
```

```python
df.head()
```

```python
df['dur_new'].astype(int).max()
```
